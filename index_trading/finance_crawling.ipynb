```python
import requests
from bs4 import BeautifulSoup
import traceback
import pandas as pd
```

```python
def parse_page(code, page):
    try:
        url = 'http://finance.naver.com/item/sise_day.nhn?code={code}&page={page}'.format(code=code, page=page)
        res = requests.get(url)
        _soap = BeautifulSoup(res.text, 'lxml')
        _df = pd.read_html(str(_soap.find("table")), header=0)[0]
        _df = _df.dropna()
        return _df
    except Exception as e:
        traceback.print_exc()
    return None
```

```python
kospi=None
url = 'http://finance.naver.com/item/sise_day.nhn?code={code}'.format(code='069500')
res = requests.get(url)
res.encoding = 'utf-8'
soap = BeautifulSoup(res.text, 'lxml')
el_table_navi = soap.find("table", class_="Nnavi")
el_td_last = el_table_navi.find("td", class_="pgRR")
pg_last = el_td_last.a.get('href').rsplit('&')[1]
pg_last = pg_last.split('=')[1]
pg_last = int(pg_last)
for page in range(1, pg_last+1):
    _df = parse_page(code='069500', page=page)
    _df = _df[['날짜','종가','시가','고가','저가']]
    _df.columns = ['kospi_date','kospi_end','kospi_start','kospi_high','kospi_low']
    kospi = pd.concat([kospi,_df])
    kospi.reset_index(drop=True, inplace=True)

kospi
```
